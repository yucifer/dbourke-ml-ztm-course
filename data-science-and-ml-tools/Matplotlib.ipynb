### Plotting Subplots

There are two options for plotting subplots.

```python
# 1st option
x = [1,2,3,4]
y = [11,22,33,44]
fig, ((ax0,ax1), (ax2,ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axis
ax0.plot(x,y)
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.hist(np.random.randn(1000));
```

```python
# 2nd option 
x = np.random.random(10)
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot to different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one to use ? (pyplot API vs Matplotlib OO API) ?
* When it comes to plotting something quickly, it is okay to use pyplot API.
* When it comes to plotting something more advanced use the matplotlib Object Oriented API.

```python
heart_disease.head()
```

```python
# Let's now plot using the OO API
# First we will create a subset of our data so we can do some data analysis as well.
over_50 = heart_disease.loc[heart_disease["age"] > 50] # we can also use heart_disease.loc[heart_disease["age"] > 50]
over_50
```

```python
# Let's try to plot a scatter plot of age v/s chol and target by using the pyplot API
over_50.plot(kind="scatter", 
            x="age",
            y="chol",
            c="target"); # 'c' stands for color. Here we want to use the color according to target column
```

```python
# Now it doesn't kinda looks good. Let's replot it using the OO method mixed with pyplot method.
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)

# Let's make it wider 
ax.set_xlim([45,100]);
```

```python
# Let's plot it using OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the Plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line at the mean cholestrol level
ax.axhline(y=over_50["chol"].mean(),linestyle='--',color="red");
```

```python
# Subplot of chol, thalach, age
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels", ylabel="Cholestrol")

# Add a legend in ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horixontal line indicating the mean cholestrol level
ax0.axhline(y=over_50.chol.mean(), linestyle='--')

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

# Add a legend in ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a horixontal line indicating the mean max heart rate level
ax1.axhline(y=over_50.thalach.mean(),linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
# Set the style of the plot
plt.style.use('seaborn-whitegrid')

# Subplot of chol, thalach, age
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter')  # cmap stands for 'color map'

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels", ylabel="Cholestrol")

# Set ax0 x_limit
ax0.set_xlim([50,80])

# Add a legend in ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horixontal line indicating the mean cholestrol level
ax0.axhline(y=over_50.chol.mean(), linestyle='--',color='red')

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap='winter')

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

# Set ax1 x_limit and y_limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend in ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a horixontal line indicating the mean max heart rate level
ax1.axhline(y=over_50.thalach.mean(),linestyle='--',color='red');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
